# Automatização de Processos

## Problema passo a passo

- Passo 1: buscar informações das ações de forma automática
- Passo 2: criar as análises dos dados
- Passo 3: enviar um e-mail para o meu gestor com os resultados das análises

## Passo 1: buscar informações das ações de forma automática

```python
!pip install yfinance
```

```python
import yfinance as yf
```

```python
ticker = input("Digite um codigo da ação: ")
dados = yf.Ticker(ticker).history("6mo")
```

```python
fechamento = dados.Close
print(fechamento)
```

```python
fechamento
fechamento.plot()
```

## Passo 2: criar as análises dos dados

- cotação máxima
- cotação mínima
- cotação atual

```python
cotacao_maxima = round(fechamento.max(), 2)
cotacao_minima = round(fechamento.min(), 2)
cotacao_atual = round(fechamento[-1], 2)

print(f"A cotação máxima é de: R${cotacao_maxima}")
print(f"A cotação mínima é de: R${cotacao_minima}")
print(f"A cotação atual é de: R${cotacao_atual}")
```

## Passo 3: enviar um e-mail para o meu gestor com os resultados das análises

- abrir uma nova aba
- digitar o endereço do GMail, ENTER
- clicar no botão "Escrever"
- digitar o destinatário,tab 
- digitar o assunto, tab
- digitar a mesangem
- clicar em enviar

```python
import pyautogui
import pyperclip
```

```python
#pause de 2 segundos entre os passos
pyautogui.PAUSE = 5

#abrir uma nova aba (ctrl+t)
pyautogui.hotkey("ctrl", "t")

#digitar o endereço do Gmail, Enter
pyperclip.copy("https://web.whatsapp.com/")
pyautogui.hotkey("ctrl","v")
pyautogui.hotkey("enter")

#buscar contato
pyautogui.hotkey("tab")
pyautogui.hotkey("tab")
pyautogui.hotkey("tab")
pyautogui.hotkey("tab")
pyautogui.hotkey("tab")
pyautogui.hotkey("tab")
pyautogui.hotkey("tab")


#buscar o contato
pyperclip.copy("Paula XXX")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")


#digitar a mensagem
mensagem = f"""
*Relatório das análises de ações*

Prezado Giovane, 

As análises das recomendações de compra de ações para o mês de Agosto estão disponíveis.

Destaco, também, os dados da empresa {ticker}, que aponta uma grande oporunidade de compra. 

Empresa:{ticker}
Cotação atual: R${cotacao_atual}

Qualquer dúvida, estou à disposição.

Att,

Giovane Seberino.
"""

pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

#enviar mensagem
pyautogui.hotkey("enter")
print("Mensagem enviada com sucesso!")
```
